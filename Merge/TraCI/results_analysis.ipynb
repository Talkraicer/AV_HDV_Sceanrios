```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_results(df, major, minor=500):
    # for each desired slow speed, plot the average number of steps with error bars
    for i in range(0,10,3):
        avg_speeds = df[f"slow_speed_{i}_avg"]
        std_speeds = df[f"slow_speed_{i}_std"]
        plt.plot(df.index, avg_speeds, label=f"slow speed {i}")
        plt.xlabel("AV percentage")
        plt.ylabel("number of steps")
        plt.title("Major/Minor Arrival Rate Per Hour: {}/{}".format(major, minor))
        plt.legend()
```

```python
plot_results(pd.read_csv("results_10_2500Major.csv",index_col=0),2500)
```
